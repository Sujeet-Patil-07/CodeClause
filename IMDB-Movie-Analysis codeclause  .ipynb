### Applying Functions

```python
def rev_cat(x):
    if x>700:
        return 'Global Blockbuster'
    elif x>500:
        return 'Super Blockbuster'
    elif x>300:
        return 'Blockbuster'
    elif x>150:
        return 'Hit'
    elif x>100:
        return 'Average'
    else:
        return 'Flop'
```

```python
rev_cat(800)
```

```python
df_imdb['Movie_Blockbuster_Category'] = df_imdb['Revenue (Millions)'].apply(rev_cat)
```

```python
#classification of the movies by rating

def rating(movie):
    if movie.Rating>=7.0:
        return "High"
    elif movie.Rating<=6.0:
        return "Good"
    else:
        return "Average"

df_imdb["Rating words"]=df_imdb.apply(rating,axis="columns") 
df_imdb["Rating words"].value_counts()
```

```python
datawords=df_imdb["Rating words"].value_counts().reset_index()
datawords=datawords.rename(columns={"index":"Rating Words","Rating words":"Count"})
datawords=datawords.set_index("Rating Words")


plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
colors=sns.color_palette('hls')
sns.barplot(x=datawords.index,y="Count",data=datawords,palette=colors)
plt.title("Rating number information")


plt.subplot(1,2,2)
sns.countplot(x="Year",hue="Rating words",data=df_imdb,palette=colors)
plt.title("Rating indicators by years")
plt.show()
```
